```python
import numpy as np
import matplotlib.pyplot as plt

m = -1
b = 1
alpha = 0.1

X = np.array([1, 3])
Y = np.array([3, 6])

m_values = [m]
b_values = [b]
errors = []

for i in range(3):
    Y_pred = m * X + b

    errors_current = Y - Y_pred

    n = len(X)
    dJ_dm = -(2/n) * np.sum(errors_current * X)
    dJ_db = -(2/n) * np.sum(errors_current)

    m = m - alpha * dJ_dm
    b = b - alpha * dJ_db

    m_values.append(round(m, 3))
    b_values.append(round(b, 3))
    mse = np.mean(errors_current**2)
    errors.append(round(mse, 3))

    print(f"Iteration {i+1}")
    print(f"Predictions: {Y_pred}")
    print(f"Errors: {errors_current}")
    print(f"dJ/dm: {dJ_dm:.3f}, dJ/db: {dJ_db:.3f}")
    print(f"Updated m: {m:.3f}, b: {b:.3f}\n")

final_Y_pred = m * X + b
print("Final Predictions:", final_Y_pred)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(len(m_values)), m_values, label="m values", marker='o')
plt.plot(range(len(b_values)), b_values, label="b values", marker='x')
plt.xlabel("Iteration")
plt.ylabel("Parameter Value")
plt.title("m and b over Iterations")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(1, 4), errors, marker='s', color='red')
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error")
plt.title("Error over Iterations")
plt.grid(True)

plt.tight_layout()
plt.show()
```
